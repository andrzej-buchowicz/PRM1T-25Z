```python
import math

print(math.sqrt(-1))
```

```python
try:
  print(math.sqrt(-1))
except ValueError as err:
  print(err)
```

```python
kod = {'A': 0, 'B': 1, 'C': 5}
print(kod['D'])
```

```python
kod = {'A': 0, 'B': 1, 'C': 5}
try:
    print(kod['D'])
except KeyError as err:
    print(type(err), err)
```

```python
kod = {'A': 0, 'B': 1, 'C': 5}   # , 'D': -1}
try:
    print(math.sqrt(kod['D']))
except KeyError as err:
    print(type(err), err)
except ValueError as err:
    print(type(err), err)
print(kod)
```

```python
kod = {'A': 0, 'B': 1, 'C': 5, 'D': -1}
try:
    print(math.sqrt(kod['D']))
except (KeyError, ValueError) as err:
    print(type(err), err)
print(kod)
```

```python
kod = {'A': 0, 'B': 1, 'C': 5}
try:
  print(kod['D'])
except KeyError as err:
  print(err.__class__.__name__, err)
finally:
  kod['D'] = 123
print(kod)
```

```python
kod = {'A': 0, 'B': 1, 'C': 5}
try:
  print(kod['D'])
except KeyError as err:
  print(err.__class__.__name__, err)
else:
  kod['D'] = 77
print(kod)
```

## Iterator

```python
class KwadratyLiczb:
  def __init__(self, start, stop):
      self.start = self.liczba = start
      self.stop = stop

  def __iter__(self):
      return self

  def __next__(self):
      liczba = self.liczba
      if liczba <= self.stop:
        self.liczba += 1
        return liczba**2                
      else:
          raise StopIteration
```

```python
kwadraty_liczb = KwadratyLiczb(1, 5)
for liczba in kwadraty_liczb:
    print(liczba)
```
