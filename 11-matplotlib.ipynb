## Pakiet `matplotlib`

### Wykres funkcji jednej zmienej

```python
import numpy as np

import matplotlib as mpl
import matplotlib.pylab as plt
```

```python
t = np.linspace(0., 1., num=100)
x = np.sin(2 * np.pi * t)
y = np.cos(2 * np.pi * t)
```

```python
_, ax = plt.subplots(figsize=(9, 6))
ax.plot(t, x, 'r-', label=r'$\sin(2 \pi t)$')
ax.plot(t, y, color='#000080', linestyle='dashed', label=r'$\cos(2 \pi t)$')
ax.grid()
ax.set_xlabel('t')
ax.set_ylabel('x(t)')
ax.set_title("Wykres funkcji")
ax.legend(loc='upper right')
#plt.show()

plt.savefig('wykres.png')
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(9, 6))

fig.suptitle("Wykres funkcji")

ax1.plot(t, x, 'r-')
ax1.grid()
ax1.set_ylabel(r'$\sin(2 \pi t)$')

ax2.plot(t, y, color='#000080', linestyle='dashed')
ax2.grid()
ax2.set_xlabel('t')
ax2.set_ylabel(r'$\cos(2 \pi t)$');
```

```python
t = np.linspace(0, 0.5, 20)
x = np.sin(2 * np.pi * t)

_, ax = plt.subplots(figsize=(9, 6))
ax.stem(t, x);
```

### Wyniki pomiarów

```python
x_pom = np.linspace(0., 10, num=10)
y_pom = np.polyval([0.5, 1.0, 2.], x_pom) + 2.5 * np.random.randn(x_pom.shape[0])
```

```python
_, ax = plt.subplots(figsize=(9, 6))
ax.plot(x_pom, y_pom, marker='d')   

ax.grid()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Wyniki pomiarów');
```

```python
p = np.polyfit(x_pom, y_pom, 2)
p
```

```python
x = np.linspace(0, 10, num=100)
y = np.polyval(p, x)
```

```python
_, ax = plt.subplots(figsize=(9, 6))

ax.scatter(x_pom, y_pom, marker='d', color='#ff0000', label='pomiar')
ax.plot(x, y, color='#ff8000', linestyle='dashed', label='aprox')

ax.grid()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
ax.set_title('Wyniki pomiarów');
```

### Wykres słupkowy

```python
labels = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 7, len(labels))
```

```python
_, ax = plt.subplots(figsize=(9, 6))
ax.bar(range(len(labels)), values, color=mpl.colormaps['Set1'].colors)
ax.set_xticks(range(len(labels)), labels);
```

```python
_, ax = plt.subplots(figsize=(9, 6))
ax.barh(range(len(labels)), values, color=mpl.colormaps['tab20'].colors)
ax.set_yticks(range(len(labels)), labels);
```

```python
data1 = np.random.randint(5, 20, len(labels))
data2 = np.random.randint(5, 20, len(labels))
```

```python
_, ax = plt.subplots(figsize=(9, 6))
ax.bar(np.arange(len(labels)) - 0.21, data1, width=0.4, color='#45a29e', label='data1')
ax.bar(np.arange(len(labels)) + 0.21, data2, width=0.4, color='#66fcf1', label='data2')
ax.legend()
ax.set_xticks(range(len(labels)), labels);
```

```python
_, ax = plt.subplots(figsize=(9, 6))
ax.bar(range(len(labels)), data1, color='#45a29e', label='data1')
ax.bar(range(len(labels)), data2, bottom=data1, color='#66fcf1', label='data2')
ax.legend()
ax.set_xticks(range(len(labels)), labels);
```

### Histogram

```python
data = np.random.randn(100_000)

_, ax = plt.subplots(figsize=(9, 6))
hist, bins_edges = np.histogram(data, bins=1_000)
# print(hist.shape, bins_edges.shape, (bins_edges[-1] - bins_edges[0]) / bins_edges.shape[0])
ax.bar(bins_edges[1:], hist, width=0.02);
```

### Wykres kolowy

```python
data3 = 100 * (data1 / np.sum(data1))

_, ax = plt.subplots(figsize=(9, 6))
ax.pie(data3, colors=mpl.colormaps['Dark2'].colors, 
       labels=labels, 
       startangle=90, counterclock=False, autopct='%1.1f%%')
ax.set_title('Wyniki')
```

### Wykres funkcji dwóch zmiennych (3D)

```python
fig, ax = plt.subplots(figsize=(9, 6), subplot_kw={"projection": "3d"})

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)

Z = X**2 - Y**2

surf = ax.plot_surface(X, Y, Z, cmap=mpl.colormaps['Spectral'],
                       linewidth=0, antialiased=False)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(elev= 30, azim = -60)
```
